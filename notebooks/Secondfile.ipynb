```python
import os

def search_for_function(root_dir, function_name):
    matches = []
    for dirpath, dirnames, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.endswith(".py"):
                full_path = os.path.join(dirpath, filename)
                try:
                    with open(full_path, "r", encoding="utf-8") as f:
                        content = f.read()
                        if function_name in content:
                            matches.append(full_path)
                except Exception as e:
                    pass
    return matches

matches = search_for_function(cabt_path, "all_card_data")
print(matches)
```

```python
SELECT_TYPE_YES_NO = 9
SELECT_TYPE_CARD = 1
SELECT_TYPE_ATTACK = 6
SELECT_TYPE_EVOLVE = 7
SELECT_TYPE_MAIN = 0

CONTEXT_IS_FIRST = 41
CONTEXT_ATTACK = 35
CONTEXT_SETUP_ACTIVE = 1
CONTEXT_SETUP_BENCH = 2

OPTION_YES = 1
OPTION_NO = 2


def decide_action(obs_dict):
    if obs_dict["select"] is None:
        return deck  # deck submission phase

    select = obs_dict["select"]
    select_type = select["type"]
    context = select["context"]
    options = select["option"]

    # Rule 1: Coin toss — always choose to go first
    if select_type == SELECT_TYPE_YES_NO and context == CONTEXT_IS_FIRST:
        for i, opt in enumerate(options):
            if opt["type"] == OPTION_YES:
                return [i]

    # Fallback for anything we haven't built logic for yet
    return list(range(select["minCount"]))


def my_agent(obs_dict):
    return decide_action(obs_dict)
```

```python
import random
from kaggle_environments import make

deck = [
    721, 721, 722, 722, 722, 722, 723, 723, 723, 723,
    1092, 1121, 1121, 1145, 1145, 1163, 1163,
    1219, 1219, 1219, 1219, 1227, 1227, 1227, 1227,
    1262, 1262,
    3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3,
    3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3
]
print("Deck length:", len(deck))

def random_agent(obs_dict):
    if obs_dict["select"] is None:
        return deck
    return random.sample(
        list(range(len(obs_dict["select"]["option"]))),
        obs_dict["select"]["maxCount"]
    )

SELECT_TYPE_YES_NO = 9
CONTEXT_IS_FIRST = 41
OPTION_YES = 1

def decide_action(obs_dict):
    if obs_dict["select"] is None:
        return deck

    select = obs_dict["select"]
    select_type = select["type"]
    context = select["context"]
    options = select["option"]

    if select_type == SELECT_TYPE_YES_NO and context == CONTEXT_IS_FIRST:
        for i, opt in enumerate(options):
            if opt["type"] == OPTION_YES:
                return [i]

    return list(range(select["minCount"]))

def my_agent(obs_dict):
    return decide_action(obs_dict)
```

```python
env = make("cabt", configuration={"episodeSteps": 200})
result = env.run([my_agent, random_agent])

print("Game finished!")
print("Final status:", [result[-1][0]["status"], result[-1][1]["status"]])
print("Final reward:", [result[-1][0]["reward"], result[-1][1]["reward"]])
```

```python
wins = 0
losses = 0
draws = 0

for game_num in range(20):
    env = make("cabt", configuration={"episodeSteps": 200})
    result = env.run([my_agent, random_agent])
    reward = result[-1][0]["reward"]
    
    if reward == 1:
        wins += 1
    elif reward == -1:
        losses += 1
    else:
        draws += 1

print(f"Wins: {wins}, Losses: {losses}, Draws: {draws}")
print(f"Win rate: {wins/20*100:.1f}%")
```

```python
def parse_energy_cost(cost_string):
    if pd.isna(cost_string):
        return []
    
    energy_map = {
        "G": "Grass",
        "R": "Fire",
        "W": "Water",
        "L": "Lightning",
        "P": "Psychic",
        "F": "Fighting",
        "D": "Darkness",
        "M": "Metal",
        "A": "Dragon",  # need to verify this one
    }
    
    costs = []
    i = 0
    while i < len(cost_string):
        char = cost_string[i]
        if char == "{":
            # find matching closing brace
            end = cost_string.index("}", i)
            symbol = cost_string[i+1:end]
            costs.append(energy_map.get(symbol, symbol))
            i = end + 1
        elif char == "●":
            costs.append("Colorless")
            i += 1
        else:
            i += 1
    return costs


# Test it on Hippowdon's Super Sandstorm attack
print(parse_energy_cost("{F}{F}●"))
print(parse_energy_cost("{D}●●"))
print(parse_energy_cost("●"))
```
